#### reviewing the basics

```python
# get the data

tips = sns.load_dataset("tips")
tips.sample(5)
```

4 basic ways to get data from a df

```python
# get columns

tips[["sex", "smoker"]].head(3)
```

```python
# get rows (but few legit use cases)
tips[3:7]
```

INSTEAD USE EITHER ILOC or LOC

```python

# .loc gets you the rows and columns based on their name (true index and column name)
tips.loc[2:4, "sex" : "smoker"]
```

```python
tips.loc[1:5, "tip": "day"] # loc is inclusive!!!!
```

```python
# .iloc does the same, but you use their ordering instead of calling them based on name

tips.iloc[1:3, 0:2]
```

```python
tips.iloc[1:5, 1:5] # .iloc is NOT inclusive!!
```

```python
# Boolean broadcasting

tips["tip"] > 5
```

```python
tips[[tips["day"] == "Sun"] and tips["smoker"] != "Yes"].head(3)
```

```python
tips[tips["tip"] > 3].head()
```

```python
tips[tips["sex"] == "Male"].tail(3)
```

```python
tips[tips["time"] != "Dinner"].head(3) 
```

#### Multi-Index

Run into this with GroupBys

```python
tips.head(3)
```

```python
mean_tips = tips.groupby(["sex", "smoker"]).agg({"tip": "mean"})
print(mean_tips)
```

```python
# wanted to double check to see if using multiple lines created an iterable df but it does not :/

test_tips = tips.groupby(["sex", "smoker"])  

test_tips = test_tips.agg({"tip":"mean"})
print(test_tips)
```

```python
mean_tips.index
```

```python
mean_tips.loc[("Male","No")]
```

#### using reset_index()

```python
test_tips
```

compare that shit show to this

this way it's much easier to select the "smoker" column

```python
ri_tips = mean_tips.reset_index()
ri_tips
```

```python
males = ri_tips[(ri_tips["smoker"] == "No") & (ri_tips["sex"] == "Male")] # gotta use & - NOT and
print(males)
```

using reset_index(level) argument 

it makes "smoker" the index instead of the automated ascending index

```python
ri0_tips = mean_tips.reset_index(level = 0)
ri0_tips
```

Rare use case: pulling indices back into the index (useful for certain types of merges)

```python
ri_tips.set_index(["sex", "smoker"])
```

```python
ri0_tips = ri0_tips.set_index("sex", append= True)
ri0_tips
```
